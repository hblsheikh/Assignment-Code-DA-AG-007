```python
#Question 1: What is hypothesis testing in statistics?

#Answer:

#Hypothesis testing is a decision-making technique in statistics that helps determine 
#whether the observed data provides enough evidence to support or reject an assumption 
#about a population. It compares sample data with a proposed claim (hypothesis) and 
#uses probability theory to conclude if the claim is likely to be true. In short, 
#it is a tool to validate assumptions using sample information.
```

```python
#Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

#Answer:

#Null Hypothesis (H₀): A neutral statement that assumes no significant effect or difference exists.
#It acts as the starting assumption.

#Alternative Hypothesis (H₁): A statement that proposes a change, effect, or difference exists in reality.

#Main Difference:
#The null hypothesis assumes stability or no effect, while the alternative hypothesis challenges that
#assumption and suggests that something new or different is happening.

#Example:

#A coin is equal chance of heads and tails
#The coin is biased (not equal chance)
```

```python
#Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

#Answer:

#The significance level (α) is the cutoff probability used to decide whether to reject the null hypothesis. 
#It represents the maximum risk of being wrong when rejecting H₀. Common values are 0.05 (5%) or 0.01 (1%).

#If the test’s p-value ≤ α, we reject H₀.
#If the p-value > α, we do not reject H₀.

#Role: The significance level controls the reliability of results and helps balance the chances of making an error during decision-making.
```

```python
#Question 4: What are Type I and Type II errors? Give examples of each.

#Answer:

#Type I Error: This occurs when we wrongly reject a true null hypothesis.

#Example: A judge declares an innocent person guilty.

#Type II Error: This occurs when we fail to reject a false null hypothesis.

#Example: A judge allows a guilty person to go free because the evidence was not strong enough.
```

```python
#Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

#Answer:

#Z-test:

#Used when the population variance or standard deviation is already known.

#Works well with large sample sizes (n > 30).

#Based on the normal distribution.

#T-test:

#Applied when the population variance is unknown.

#Better for smaller sample sizes (n ≤ 30).

#Uses the t-distribution, which adjusts for uncertainty with small data.

#When to use:

#Z-test → known variance, large samples.

#T-test → unknown variance, small samples.
```

```python
#Question 6: Python Program – Binomial Distribution Histogram

#Answer:

import numpy as np
import matplotlib.pyplot as plt

n = 10   # number of trials
p = 0.5  # probability of success

# Generate binomial distribution
data = np.random.binomial(n, p, 1000)

plt.hist(data, bins=range(n+2), edgecolor='black', density=True)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

```python
#Question 7: Hypothesis Testing using Z-statistics in Python

#Answer:

import numpy as np
from scipy import stats

sample_data = [
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
]

mu = 50

mean_sample = np.mean(sample_data)
std_sample = np.std(sample_data, ddof=1)  # sample standard deviation
n = len(sample_data)

z_stat = (mean_sample - mu) / (std_sample / np.sqrt(n))

p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Size (n):", n)
print("Sample Mean:", round(mean_sample, 3))
print("Sample Std Dev:", round(std_sample, 3))
print("Z-statistic:", round(z_stat, 3))
print("p-value:", round(p_value, 4))

alpha = 0.05
if p_value < alpha:
    print("Reject the null (difference).")
else:
    print("Fail to reject(no difference).")
```

```python
#Question 8: Python Script – Normal Distribution & 95% Confidence Interval

#Answer:

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

data = np.random.normal(loc=50, scale=5, size=1000)

mean = np.mean(data)
se = stats.sem(data)

ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=se)

print("Mean:", mean)
print("95% Confidence Interval:", ci)

plt.hist(data, bins=30, edgecolor='black', density=True)
plt.title("Normal Distribution with 95% CI")
plt.axvline(ci[0], color='red', linestyle='dashed', linewidth=2)
plt.axvline(ci[1], color='red', linestyle='dashed', linewidth=2)
plt.show()
```

```python
#Question 9: Python Function – Z-scores & Histogram

#Answer:

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

data = np.random.normal(100, 15, 1000)

z_scores = zscore(data)

def plot_zscores(z_scores):
    plt.hist(z_scores, bins=30, edgecolor='black', density=True)
    plt.title("Histogram of Z-scores")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.show()

plot_zscores(z_scores)

print("First 10 Z-scores:", z_scores[:10])
```
